```python
df_total_users = pd.DataFrame(df_user_gameplay.groupby('user_id')['games_played'].sum()).reset_index() #making a new df total users grouped on the basis of user_id and total number of games played for each user.
# reset_index() makes the user_id a column instead of keeping it as an index
df_total_users #checking
```

```python
df_total_deposits = pd.DataFrame(df_deposit.groupby('user_id')['deposit_amount'].sum()).reset_index() #making a new df total deposits grouped on the basis of user_id and total amount deposited by each user.
df_total_deposits #checking
```

```python
df_total_withdrawals = pd.DataFrame(df_withdrawal.groupby('user_id')['withdrawal_amount'].sum()).reset_index() #making a new df total withdrawals grouped on the basis of user_id and total amount withdrawn by each user.
df_total_withdrawals #checking
```

```python
# Merging the Respective Data Frames to Create a Final Data Frame
df_merged = pd.merge(pd.merge(df_total_users,df_total_deposits,on='user_id'),df_total_withdrawals,on='user_id')
# merged total users and total deposits on user id then merged the resultant with total withdrawals on user id to form a final merged df
df_merged #checking
```

Final Loyalty Points Formula:

Loyalty Point = (0.01 * Deposit) + (0.005 * Withdrawal Amount) + (0.001*(Maximum of (deposit - withdrawal))) + (0.2 * Number Of Games Played)

```python
# Creating a column for loyalty points based on the formula mentioned above
df_merged['loyalty_points'] = (0.01 * df_merged['deposit_amount']) + (0.005 * df_merged['withdrawal_amount']) + (0.001 * abs(df_merged['deposit_amount'] - df_merged['withdrawal_amount'])) + (0.2 * df_merged['games_played'])
df_merged #checking
```

## Part B - How much bonus should be allocated to leaderboard players?

After calculating the loyalty points for the whole month find out which 50 players are at the top of the leaderboard. The company has allocated a pool of Rs 50000 to be given away as bonus money to the loyal players. Now the company needs to determine how much bonus money should be given to the players. Should they base it on the amount of loyalty points? Should it be based on number of games? or Something else?

Suggest a suitable way to divide the allocated money keeping in mind the following points:
1. Only top 50 ranked players are awareded bonus

```python
df_merged.head() #checking
```

```python
df_partb = df_merged # storing it in a new data frame to avoid changes in the df_merged data frame
df_partb.head() #checking
```

```python
df_partb = df_partb.sort_values(by='loyalty_points', ascending=False) # Sorted the 'loyalty_points' column in descending order
df_partb.head() #checking
```

```python
df_partb['rank'] = df_partb.sort_values(by=['loyalty_points', 'games_played'], ascending = [False, False]).reset_index().index + 1 # ranking
df_partb.head(50) # top 50
```

- *I think the bonus money should be given to the players based on the amount of loyalty points because this is the only feature which carries an optimal weightage. (Loyalty Points have the weightage of the features games played, deposits and withdrawals)*

```python
df_partb['loyalty_points'].sum() # total loyalty points Overall
```

```python
df_partb_top50 = df_partb.head(50) # top 50 players
df_partb_top50 #checking
```

```python
df_partb_top50['loyalty_points'].sum() # total loyalty points of top 50 players
```

So the Total Loyalty Point of Top 50 Players is 653257.387

As we are distributing the 50000 prize pool among the top 50 players in the leaderboard on the basis of their loyalty points, So we have to divide 50000 by Total Loyalty Point and multiply it by the individual loyalty points.

Therefore, *Bonus = Loyalty Points * (50000 / Total Loyalty Point)*

Example: Bonus ( rank 1) = 99066.030 * (50000 / 653257.387)
                         = 99066.030 * 0.0765395095333
                         = 7582.4653476134

```python
df_partb_top50['bonus'] = df_partb_top50['loyalty_points'] * (50000 / 653257.387) # calculating the bonus
df_partb_top50 #checking
```

```python
# Data Visualization using barplot
plt.figure(figsize=(12, 6)) # To avoid overriding texts
sns.barplot(
    data=df_partb_top50,
    x='rank', # x-axis
    y='bonus', # y-axis
    palette='viridis' # Giving Gradient Graphics
)

#Labels
plt.xlabel('Player Rank', fontsize=10)
plt.ylabel('Bonus Amount', fontsize=10)

# Display the plot
plt.tight_layout() # Enhances the visuals
plt.show()
```

Hence, the top 50 players in the leaderboard received bonuses on the basis of their loyalty points.
